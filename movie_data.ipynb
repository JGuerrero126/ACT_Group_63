```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in the csv files and save them as variables

credits_data = "Resources/credits.csv"
titles_data = "Resources/titles.csv"

#create data frames using the newly created variables

credits_data_df = pd.read_csv(credits_data)
titles_data_df = pd.read_csv(titles_data)
```

```python
#remove special characters in the genres column

spec_chars = ["!",'"',"#","%","&","'","(",")",
              "*","+",",","-",".","/",":",";","<",
              "=",">","?","@","[","\\","]","^","_",
              "`","{","|","}","~","â€“"]

for char in spec_chars:
    titles_data_df['genres'] = titles_data_df['genres'].str.replace(char, ' ')
```

```python
zx_titles_data = "Resources/zx_titles.csv"

zx_titles_df = pd.read_csv(zx_titles_data, encoding='latin-1')

zx_titles_df
```

```python
#Turn Genre List into a DataFrame for easy viewing to complete next part

unique_genre_list = zx_titles_df['genres'].unique()

genre_list_df = pd.DataFrame(unique_genre_list)
genre_list_df
```

```python
#Create filters to separate the movies by genre

#Action Genre DataFrame
only_action_df = zx_titles_df.loc[zx_titles_df["genres"] == "action", :]

#Family Genre DataFrame
only_family_df = zx_titles_df.loc[zx_titles_df["genres"] == "family", :]

#Animation Genre DataFrame
only_animation_df = zx_titles_df.loc[zx_titles_df["genres"] == "animation", :]

#History Genre DataFrame
only_history_df = zx_titles_df.loc[zx_titles_df["genres"] == "history", :]

#Scifi Genre DataFrame
only_scifi_df = zx_titles_df.loc[zx_titles_df["genres"] == "scifi", :]

#Romance Genre DataFrame
only_romance_df = zx_titles_df.loc[zx_titles_df["genres"] == "romance", :]

#Comedy Genre DataFrame
only_comedy_df = zx_titles_df.loc[zx_titles_df["genres"] == "comedy", :]

#Fantasy Genre DataFrame
only_fantasy_df = zx_titles_df.loc[zx_titles_df["genres"] == "fantasy", :]

#Horror Genre DataFrame
only_horror_df = zx_titles_df.loc[zx_titles_df["genres"] == "horror", :]

#Drama Genre DataFrame
only_drama_df = zx_titles_df.loc[zx_titles_df["genres"] == "drama", :]

#Thriller Genre DataFrame
only_thriller_df = zx_titles_df.loc[zx_titles_df["genres"] == "thriller", :]

#Music Genre DataFrame
only_music_df = zx_titles_df.loc[zx_titles_df["genres"] == "music", :]

#Scifi Genre DataFrame
only_sport_df = zx_titles_df.loc[zx_titles_df["genres"] == "sport", :]

#Western Genre DataFrame
only_western_df = zx_titles_df.loc[zx_titles_df["genres"] == "western", :]

#European Genre DataFrame
only_european_df = zx_titles_df.loc[zx_titles_df["genres"] == "european", :]

#War Genre DataFrame
only_war_df = zx_titles_df.loc[zx_titles_df["genres"] == "war", :]
```

```python
only_action_df
```

```python
#Store Averages for each Movie Genre imdb_score in unique variables

action_average = only_action_df["imdb_score"].mean()
print(float(action_average))
```

```python
#Calculate Averages for Each Genre

genre_group = zx_titles_df.groupby(["genres"])

g_imdb_scores_averages = genre_group.mean()

g_imdb_scores_averages_df = pd.DataFrame(g_imdb_scores_averages)

g_imdb_scores_averages_df
```

```python
x_axis = np.arange(len(g_imdb_scores_averages_df))
tick_locations = [value for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(x_axis, g_imdb_scores_averages_df["imdb_score"], color='r', alpha=0.5, align="center")
plt.xticks(tick_locations, genre_group, rotation="vertical")
```
